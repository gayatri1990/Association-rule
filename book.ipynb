```python
dt = df
for i in df.columns:
    dt[i]=dt[i].replace(1,i)
    dt[i]=dt[i].replace(0,'NULL')
dt.head()   
```

```python
records = []
for i in range(0,2000):
    records.append([str(dt.values[i,j])for j in range(0,11)])
len(records)  
```

```python
rules = apriori(records,min_support = 0.005,min_confidence = 0.04,min_lift =3,min_length = 2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results ]
    lifts = [result[2][0][3] for result in results ]
    return list(zip(lhs,rhs,supports,confidences,lifts))
resultsinDataFrame = pd.DataFrame(inspect(results),columns = ['Left Hand Side','Right Hand Side','Support','Confidence','Lifts'])
```

```python
resultsinDataFrame.nlargest(n=10,columns = 'Lifts')
```
